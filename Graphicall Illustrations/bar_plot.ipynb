```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
d = {'WC-CB': [2.27,2.40 ], 'AdaBoost': [10.8,10.9],'BiLSTM': [24.8,40.7],'GRU': [34.2,49.1]}
data=pd.DataFrame(data=d, index=['Golden', 'Johannesburg'])
print(data)
x1=data['WC-CB']
x2=data['AdaBoost']
x3=data['BiLSTM']
x4=data['GRU']
y=["Golden","Johannesburg"]
sns.set_style("white")
y_axis=np.arange(len(y))
hatches = ['*', '.', 'x', '/']
plt.bar(y_axis, x1,0.2, label= 'WC-CB' ,edgecolor='black', color=['silver'],hatch=hatches[0])
plt.bar(y_axis+0.2, x2,0.2, label= 'AdaBoost' ,edgecolor='black', color=['royalblue'],hatch=hatches[1])
plt.bar(y_axis+0.4, x3,0.2, label= 'BiLSTM',edgecolor='black',color=['sandybrown'],hatch=hatches[2])
plt.bar(y_axis+0.6, x4,0.2, label= 'GRU',edgecolor='black', color=['darkseagreen'],hatch=hatches[3])
plt.xticks(y_axis, y)
plt.grid(color = 'black', linestyle = '--', linewidth = 0.2)
#plt.xlabel("Years")
plt.ylabel("Avg RMSE (Watt per meter square)")
#plt.rcParams["figure.figsize"] = (4, 4)
plt.legend()
plt.savefig('barplot.pdf') 
plt.savefig('barplot.eps') 
plt.savefig('barplot.png') 
plt.savefig('barplot.svg')
```

```python
import seaborn as sns
import numpy as np
d = {'WC-CB': [2.27,2.40 ], 'CatBoost': [2.71,3.94]}
data=pd.DataFrame(data=d, index=['Golden', 'CatBoost'])
print(data)
x1=data['WC-CB']
x2=data['CatBoost']
y=["Golden","Johannesburg"]
sns.set_style("white")
y_axis=np.arange(len(y))
hatches = ['.', 'x', '/']
plt.bar(y_axis, x1,0.2, label= 'WC-CB' ,edgecolor='black', color=['silver'],hatch=hatches[0])
plt.bar(y_axis+0.2, x2,0.2, label= 'CatBoost' ,edgecolor='black', color=['royalblue'],hatch=hatches[1])
plt.xticks(y_axis, y)
plt.grid(color = 'black', linestyle = '--', linewidth = 0.2)
#plt.xlabel("Years")
plt.ylabel("RMSE (Watt per meter square)")
#plt.rcParams["figure.figsize"] = (4, 1)
plt.legend()
plt.savefig('barplot_hybrid_conventional.pdf') 
plt.savefig('barplot_hybrid_conventional.eps') 
plt.savefig('barplot_hybrid_conventional.png')
plt.savefig('barplot_hybrid_conventional.svg')
```
