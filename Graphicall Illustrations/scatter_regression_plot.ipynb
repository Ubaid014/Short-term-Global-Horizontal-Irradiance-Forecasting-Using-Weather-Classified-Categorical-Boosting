```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

###GOLDEN CLOUDY

```python
df=pd.read_csv("Golden_cloudy.csv")
df.head()
```

```python
features=list(df)
features
```

```python
x=df["Actual"]
y=df["Prediction"]
```

```python

scat=sns.regplot(x=x, y=y, data=df,line_kws={"color":"orange","alpha":1,"lw":3})
#plt.title("Golden",fontdict= {"fontweight":'bold'})
plt.rcParams['font.size'] = '16'
#plt.rcParams["figure.figsize"] = (8, 6)
plt.rcParams['font.size'] = '12'
plt.grid(color = 'black', linestyle = '--', linewidth = 0.2)
plt.xlabel("Measured GHI")
plt.ylabel("Predicted GHI")
plt.savefig('Golden_reg_cloudy.pdf',bbox_inches='tight') 
plt.savefig('Golden_reg_cloudy.eps',bbox_inches='tight') 
plt.savefig('Golden_reg_cloudy.png',bbox_inches='tight') 
plt.show()
```
