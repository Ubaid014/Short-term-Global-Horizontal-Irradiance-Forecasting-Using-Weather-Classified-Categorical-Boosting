#######Golden cloudy

```python
import pandas as pd
import matplotlib.pyplot as plt
df= pd.read_csv("Golden_cloudy.csv")
features=list(df)
print(features)
plt.rcParams['font.size'] = '16'
plt.rcParams["figure.figsize"] = (8, 6)
plt.plot(df['Actual'], label= 'Measured',marker = ".",)
plt.plot(df['CatBoost'],label='WC-CB' ,marker = '+',)
plt.plot(df['AdaBoost'],label='AdaBoost',marker = 'o', )
plt.plot(df['BiLSTM'],label='BiLSTM',marker = '*', )
plt.plot(df['GRU'],label='GRU',marker = "s", )
plt.grid(color = 'black', linestyle = '--', linewidth = 0.2)
plt.xlabel('Time (h)')
#plt.title(" Graphical comparison")
plt.ylabel('GHI (watt per meter square)')
plt.legend(prop={"size":12})
plt.savefig('Golden_cloudy.pdf',bbox_inches='tight')
plt.savefig('Golden_cloudy.eps',bbox_inches='tight')
plt.savefig('Golden_cloudy.png',bbox_inches='tight')
plt.savefig('Golden_cloudy.svg',bbox_inches='tight')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1.inset_locator import inset_axes

# Load the data
df = pd.read_csv("Golden_cloudy.csv")
features = list(df)

# Set up the main plot
plt.rcParams['font.size'] = '16'
plt.rcParams["figure.figsize"] = (8, 6)

# Plot the main data
plt.plot(df['Actual'], label='Measured', linestyle='--',marker=".")
plt.plot(df['CatBoost'], label='WC-CB',linestyle='--', marker='+')
plt.plot(df['AdaBoost'], label='AdaBoost',linestyle='--', marker='o')
plt.plot(df['BiLSTM'], label='BiLSTM',linestyle='--', marker='*')
plt.plot(df['GRU'], label='GRU',linestyle='--', marker="s")

plt.grid(color='black', linestyle='-', linewidth=0.2)
plt.xlabel('Time (h)')
plt.ylabel('GHI (watt per meter square)')
plt.legend(prop={"size": 12}, loc='lower right')

# Identify the points for the zoomed-in section
zoom_start = 11
zoom_end = 13

# Define the position and size of the inset
axins = inset_axes(plt.gca(), width="50%", height="30%", loc='upper right')

# Plot the zoomed-in data in the inset
axins.plot(df['Actual'][zoom_start:zoom_end + 1],linestyle='--', label='Measured', marker=".")
axins.plot(df['CatBoost'][zoom_start:zoom_end + 1], label='WC-CB',linestyle='--', marker='+')
axins.plot(df['AdaBoost'][zoom_start:zoom_end + 1], label='AdaBoost',linestyle='--', marker='o')
axins.plot(df['BiLSTM'][zoom_start:zoom_end + 1], label='BiLSTM',linestyle='--', marker='*')
axins.plot(df['GRU'][zoom_start:zoom_end + 1], label='GRU',linestyle='--', marker="s")

# Customize and annotate the inset
#axins.set_xlabel('Time (h)')
#axins.set_ylabel('GHI (watt per meter square)')
#axins.set_title('Zoomed-In View')
#axins.legend(prop={"size": 8})
axins.tick_params(axis='x', labelsize=10)  # Adjust the font size as needed
axins.tick_params(axis='y', labelsize=10)
# Connect the main graph and inset with lines or arrows
plt.annotate("",(zoom_start, df['Actual'][zoom_start]), xytext=(zoom_start + 1, df['Actual'][zoom_start] + 50),
             )

plt.savefig('Golden_cloudy.pdf',bbox_inches='tight')
plt.savefig('Golden_cloudy.eps',bbox_inches='tight')
plt.savefig('Golden_cloudy.png',bbox_inches='tight')
plt.savefig('Golden_cloudy.svg',bbox_inches='tight')

# Show the plot
plt.show()
```
